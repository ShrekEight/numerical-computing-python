# singed and unsigned integers

```
number of bits    type      min      max        numpy type name
       8        unsigned     0       255         uint8
       8          signed    -128     127          int8
```

````python
import numpy as np
````

````python
x = np.int8(3)
x
````

````python
type(x)
````

````python
x.itemsize   # number of bytes
````

````python
x2 = np.int16(3)
x2
````

````python
type(x2)
````

````python
x2.itemsize   # number of bytes
````

````python
np.arange(0, 128, dtype=np.int8)  # ok
````

````python
np.arange(0, 256, dtype=np.int8)  # integer overflow!
````

````python
np.arange(0, 256, dtype=np.uint8)  # ok
````

````python
np.arange(0, 256, dtype=np.int16) 
````
